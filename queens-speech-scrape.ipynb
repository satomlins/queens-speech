```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
import dtale as dt
```

```python
import re
import string
from collections import Counter 
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
# import nltk
# nltk.download('stopwords')
# nltk.download('punkt')
```

```python
all_stopwords = stopwords.words('english')
all_stopwords = [a.lower() for a in all_stopwords]
all_stopwords.extend(['i', 'us','we'])
```

```python
translator = str.maketrans('', '', string.punctuation)
```

```python
def get_text_from_url(url):
    
    # stores the HTML page in 'soup', a BeautifulSoup object
    return BeautifulSoup(requests.get(url).content).get_text()
```

```python
speech = {}
page_names = ['christmas-broadcast-','queenâ€™s-christmas-broadcast-', 'queens-christmas-broadcast']

for year in range(1953, 2020):
    
    for page_name in page_names:
        url = 'https://www.royal.uk/' + page_name + str(year)
        web_text = get_text_from_url(url)
    
        if 'Page not found' not in web_text:
            break
        
        if page_name == page_names[-1]:
            web_text = ''
            
    curr_speech = re.search('Christmas Broadcast [0-9]{4}\n\n\n\n\n ((.|\n)*)Share this article:', web_text).group(1)[26:].lower()
    curr_speech = curr_speech.translate(translator)
    speech[year] = [[word for word in word_tokenize(curr_speech) if not word.lower() in all_stopwords]]
```

```python
speech_df = pd.DataFrame.from_dict(speech, orient='index', columns=['text'])

speech_df.head()
```

```python
results_df = speech_df.drop('text', axis=1)

for i in range(20):

    results_df['common_words_' + str(i)] = speech_df['text'].apply(lambda x: Counter(x).most_common(20)[i])

results_df.head()
```

```python
dt.show(results_df, ignore_duplicate=True).open_browser()
```
